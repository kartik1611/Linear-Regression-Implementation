```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import os
os.chdir("F://buffalo//buffalo project//linear regression from scratch")
```

```python
X=[]
Y=[]

for i in open("data_1d.csv"):
    x,y =  i.split(",")
    X.append(float(x))
    Y.append(float(y))
```

# converting to numpy array

```python
X=np.array(X)
Y=np.array(Y)
```

# plot 

```python
plt.scatter(X,Y)
plt.show
```

# applying the mathematical equation and calculating a and b

```python
denom=X.dot(X)-X.mean()*X.sum()
a=(X.dot(Y)-Y.mean()*X.sum())/denom
b=(Y.mean()*X.dot(X)-X.mean()*Y.dot(X))/denom
```

# Calculation predicted Y 

```python
yhat= a*X+b
```

```python
plt.scatter(X,Y)
plt.plot(X,yhat)
plt.show()
```

 Calculating R square

```python
d1 =(Y-yhat)
d2=(Y-Y.mean())

rsquare=1-(d1.dot(d1))/(d2.dot(d2))
print("rsquare is ",rsquare)
```

## multiple linear regression

```python
X=[]
Y=[]
for i in open("data_2d.csv"):
    x1,x2,y =  i.split(",")
    X.append([float(x1),float(x2),1])
    Y.append(float(y))
```

converting into numpy array

```python
X=np.array(X)
Y=np.array(Y)
```

 plotting data

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
#ax.scatter([X:0],[X:1],Y)
ax.scatter(X[:,0],X[:,1],Y)

plt.show()
```

# calculating weights of the model

```python
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
yhat = np.dot(X, w)            
```

```python
d1 =(Y-yhat)
d2=(Y-Y.mean())

rsquare=1-(d1.dot(d1))/(d2.dot(d2))
print("rsquare is ",rsquare)
```

# Polynomial Regression

```python
X=[]
Y=[]
for i in open("data_poly.csv"):
    x,y =  i.split(",")
    x=float(x)
    X.append([1,x,x*x])
    Y.append(float(y))
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
# plotting
plt.scatter(X[:,1],Y)
plt.show()
```

```python
# calculate weights
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
yhat = np.dot(X, w)   
```

```python
X2only=df[['X2','ones']]
X3only=df[['X3','ones']]
```

```python
def get_r2(X,Y):
    w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
    yhat = np.dot(X, w) 
    
    d1=Y-yhat
    d2=Y-Y.mean()
    r2=1-d1.dot(d1)/d2.dot(d2)
    return (r2)
```
